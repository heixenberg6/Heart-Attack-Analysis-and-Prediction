#### 1.1 Examining the Project Topic <a id=2></a>

##### What is a Heart Attack?

* A myocardial infarction (commonly called a heart attack) is an extremely dangerous condition caused by a lack of blood flow to your heart muscle. 
* The lack of blood flow can occur because of many different factors but is usually related to a blockage in one or more of your heart’s arteries. 
* Without blood flow, the affected heart muscle will begin to die. If blood flow isn’t restored quickly, a heart attack can cause permanent heart damage and death.

#### How does a heart attack occur?


* A heart attack occurs when the flow of blood to the heart is severely reduced or blocked. The blockage is usually due to a buildup of fat, cholesterol and other substances in the heart (coronary) arteries. 
* The fatty, cholesterol-containing deposits are called plaques. The process of plaque buildup is called atherosclerosis.
* Sometimes, a plaque can rupture and form a clot that blocks blood flow. A lack of blood flow can damage or destroy part of the heart muscle.

#### What are the symptoms of heart attack?

* <b>Chest pain or discomfort.</b> Most heart attacks involve discomfort in the center or left side of the chest that lasts for more than a few minutes or that goes away and comes back. The discomfort can feel like uncomfortable pressure, squeezing, fullness, or pain.
* <b>Feeling weak, light-headed, or faint.</b> You may also break out into a cold sweat.
* <b>Pain or discomfort in the jaw, neck, or back.</b>
* <b>Pain or discomfort in one or both arms or shoulders.</b>
* <b>Shortness of breath.</b> This often comes along with chest discomfort, but shortness of breath also can happen before chest discomfort.

```python
from scipy import stats
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

```python

          
z_scores_trtbps = zscore(df["trtbps"])
for threshold in range(1, 4):
    print("Threshold Value: {}".format(threshold))
    print("Number of Outliers: {}".format(len(np.where(z_scores_trtbps > threshold)[0])))
    print("-------------------")
   
```

This variable was close to having a normal distribution despite having outliers. For this to find outliers we use zscore method. Zscore standard deviation deviates from mean in a normal (or near normal)distribution.
In a normal distribution 68% observations are one standard deviation away from the mean. 95% are 2 SD away. 99.7% are 3 SD away.   


Generally observations 3 SD away can be 

```python
z_scores_trtbps=zscore(df["trtbps"])
for threshold in range(1,4):
    print("Threshold Value: {}".format(threshold))
    print("Number of Outliers: {}".format(len(np.where(z_scores_trtbps> threshold)[0])))
    print("-------------")
```

If we keep threshold value as 1 there will be too many outliers. We can set threshold as 2.

What we can do about outliers:
a) Delete(can affect volatility of data, should be avoided)
b) Value Assignment(could be any fixed value)
c) Limitation (special assignment)
In this case, deletion not needed since outliers are very few.
Assignment also not good here, we try limitation.


```python
df[z_scores_trtbps>2].trtbps.min()
```

```python
df[df["trtbps"]<170].trtbps.max()
```

This finds the highest value lower than the lowest value of the outliers. And now we can limit the outliers with the help of this value.

Winsorire takes two values, first is the % value of the value we want to limit from the bottom of the variable. Second is the % value from top of the variable.

Here we need to find 165 corresponds to what % of the whole variable.

```python
winsorize_percentile_trtbps=(stats.percentileofscore(df["trtbps"],165))/100
print(winsorize_percentile_trtbps)
```

```python
trtbps_winsorize= winsorize(df.trtbps,(0,(1-winsorize_percentile_trtbps)))
```

0 because we dont want to make any limitations from the bottom. Other value is the value we found above, that is the upper limit.

 #### 5.2.2.2 thalach variable <a id= 32></a> 

When we visualized outliers in earlier section, we noticed that there is a slight left skew(not normal distribution). Here we cant use z score method. Instead we wil use Interquartile range method. (IQR)

IQR is the difference between 75% and 25% values of the dataset, it represents the middle 50% of data.

So , values  1.5 times less than 25% quartile and 1.5 times more than 75% quartile will be considered outliers.



```python
def iqr(df,var):
    q1= np.quantile(df[var],0.25)
    q3= np.quantile(df[var],0.75)
    diff=q3-q1
    lower_v= q1-(1.5*diff)
    upper_v=q3+(1.5*diff)
    return df[(df[var]<lower_v)|(df[var]>upper_v)]
```

```python
thalach_out=iqr(df,"thalach")
```

 #### 6.1.2 ROC curve and area under the curve <a id= 42></a> 

We continue to measure model performance.

ROC curve the rock curve is a very important measure for classification problems.

Rock is a probability curve. The higher a model we create is above the dashed line the better it is. 
We shouldnt use models that are under the dashed line. 

Y axis is true positive ratio-> recall sensitivity values, rate of detecting people who are likely to have a heart attack.

X axis-> false positive rate, rate at which we make an incorrect prediction. Rock curve is formed at the point where the  TPR and FPR values intersect.

AUC-> Area under the curve tells us how well the model can distinguish the classes. Higher the AUC better the model at detecting 0s as 0s  and 1s as 1s.

Reference->

AUC Values       Test Quality
0.9-1.0          Excellent
0.8-0.9          Very Good
0.7-0.8          Good
0.6-0.7          Satisfactory
0.5-0.6          Unsatisfactory


We use these even though we have checked accuracy and cross validated it, because it will be better for our model and accuracy alone will not be enough.


```python
from sklearn.metrics import plot_roc_curve
```
